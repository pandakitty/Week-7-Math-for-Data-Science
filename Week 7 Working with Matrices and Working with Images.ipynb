```python
#  1. Enter each of the matrices A, B, C as Numpy matrix
```

```python
# import numpy
import numpy as np

# create a matrix from a string

A = np.matrix ('1, -2, 3, 7; 2, 1, 1, 4; -3, 3, -2, 10')

B = np.matrix ('0, 1, -3, -2; 10, -1, 2, -3; 5, 1, -1, 4')

C = np.matrix ('4, 0, -2, 3; 3, 6, 9, 7; 2, 2, 5, 1; 9, 4, 6, -2')

m,n = A.shape
o,p = B.shape
q,r = C.shape

# print 

print("Matrix A: ", A)
print ("A has %d rows and %d columns" %(m,n))

print("Matrix B: ", B)
print ("B has %d rows and %d columns" %(o,p))

print("Matrix C: ", C)
print ("C has %d rows and %d columns" %(q,r))
```
